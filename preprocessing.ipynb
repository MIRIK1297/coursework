Cut API-calls traces from cuckoo logs from dataset (https://zenodo.org/record/1203289#.XtaXUmhRXmh)

```python
input_dir = './data/cuckoo/'
output_dir = './data/preprocessed/'

import os
import json

for dirname in os.listdir(input_dir):
    print(dirname)
    processed_files = 0
    for proc_dir in os.listdir(input_dir + dirname):
        filepath = input_dir + dirname + '/' + proc_dir + '/reports/report.json'
        if os.path.exists(filepath):
            try:
                with open(filepath) as json_file:
                    procs_traces = list(filter(lambda proc: proc['calls'] != [],json.load(json_file)['behavior']['processes']))

                procs_traces = list(map(lambda proc: {'name': proc['process_name'], 'calls':
                                                      list(map(lambda call: { call['api']: list(call['arguments'].values())},
                                                              proc['calls']))}, procs_traces))
                if procs_traces:
                    if not os.path.exists(output_dir + dirname + '/'):
                        os.makedirs(output_dir + dirname + '/')
                    with open(output_dir + dirname + '/' + proc_dir + '.json', 'w') as out_file:
                        json.dump(procs_traces, out_file)

                processed_files += 1
                if processed_files % 10 == 0:
                    print("processed files: {}".format(processed_files))
            except:
                print('problems with: ' + filepath)
            
```
